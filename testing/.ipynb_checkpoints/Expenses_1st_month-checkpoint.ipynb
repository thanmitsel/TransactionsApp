# Testing Bokeh

```python
from bokeh.plotting import figure
from bokeh.io import show, output_file, output_notebook

# Import the ColumnDataSource class
from bokeh.models import ColumnDataSource, HoverTool

import math

output_file("bar_sorted.html")

# Define the parameters for plotting
x_name = 'Category'
measure = 'Amount'
y_name = 'Total'
t_count = 'count'

grp = trans.groupby(x_name)[measure].agg(['sum','count']).reset_index().rename(columns={'sum': y_name}).sort_values(by = y_name, ascending = False)

# Convert dataframe to column data source
src = ColumnDataSource(data = grp)
# src.data.keys()

# Create the blank plot
p = figure(x_range = grp[x_name], plot_width = 600, plot_height=600,
           title= x_name + ' vs ' + y_name, 
           x_axis_label = x_name, y_axis_label = y_name)

p.vbar(source = src, x = x_name, 
       bottom = 0, top = y_name, width = 0.9,
       fill_color='grey', line_color='black',fill_alpha = 0.75,
       hover_fill_alpha = 1.0, hover_fill_color = 'green')

p.xaxis.major_label_orientation = math.pi/2

# Hover tool referring to our own data field using @ and
# a position on the graph using $
h = HoverTool(tooltips = [('Amount Spent', '@'+y_name),
                         ('Count of transactions', '@'+t_count)] )

# Style the plot
#p = style(p)

# Add the hover tool to the graph
p.add_tools(h)

show(p)
```
